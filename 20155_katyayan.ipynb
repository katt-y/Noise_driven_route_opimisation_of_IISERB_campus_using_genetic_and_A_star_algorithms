## Defining the important functions

```python
def reproduce(x,y):
    k = np.random.randint(1,100)
    if k <= 50:
        z = x[:len(x)//2] + y[len(y)//2:]
        return z

    else:
        z = y[:len(x)//2] + x[len(y)//2:]
        return z


def get_distance(x,y):
    
    distance = (great_circle(x, y).km)*1000
    return distance

def distance_bw_neighbours(x,y):

     if dict_of_nodes[x].neighbours[y] != 0:
         list_of_points = dict_of_nodes[x].neighbours[y] 
         l = [(list_of_points[i],list_of_points[i+1]) for i in range(len(list_of_points) - 1)]
         distance = 0

         for i in l:
             distance += get_distance(i[0],i[1])
         
         return distance


     elif  dict_of_nodes[y].neighbours[x] != 0:
         list_of_points = dict_of_nodes[y].neighbours[x]
         l = [(list_of_points[i],list_of_points[i+1]) for i in range(len(list_of_points) - 1)]
         distance = 0

         for i in l:
             distance += get_distance(i[0],i[1])
         
         return distance
         
     
     else:
         return get_distance(dict_of_nodes[x].on_road_coordinates,dict_of_nodes[y].on_road_coordinates)
         
def distance_bw_neighbours_2(x,y):
     cloned_dict = copy.deepcopy(dict_of_nodes)
     for i in dict_of_nodes:
        l = list(dict_of_nodes[i].neighbours.keys())
        for j in range(1,24):
            if j not in l:
                cloned_dict[i].neighbours[j] = 0
     
     if cloned_dict[x].neighbours[y] != 0:
         list_of_points = cloned_dict[x].neighbours[y] 
         l = [(list_of_points[i],list_of_points[i+1]) for i in range(len(list_of_points) - 1)]
         distance = 0

         for i in l:
             distance += get_distance(i[0],i[1])
         
         return distance


     elif  cloned_dict[y].neighbours[x] != 0:
         list_of_points = cloned_dict[y].neighbours[x]
         l = [(list_of_points[i],list_of_points[i+1]) for i in range(len(list_of_points) - 1)]
         distance = 0

         for i in l:
             distance += get_distance(i[0],i[1])
         
         return distance
         
     
     else:
         return get_distance(cloned_dict[x].on_road_coordinates,cloned_dict[y].on_road_coordinates)
         
def array_cost(array):
    cost = 0
    array_2 = [(array[i],array[i+1]) for i in range(len(array)-1)] 
    array_3 = np.array([distance_bw_neighbours_2(x[0],x[1]) for x in array_2])
    array_4 = np.array([noise_dict[i]*100 for i in array])
    cost = np.sum(array_3) + np.sum(array_4)
    return cost       
                

def compute_route(points, A_star = True, genetic = False):

    if A_star == True:

        start,destination = (points[0],points[1])
        set_of_coordinates = []
        pointer = start
        cost = 0
    
        while pointer != destination:
           # set_of_coordinates.pop()
            l = [i for i in dict_of_nodes[pointer].neighbours]
            g_heuristic_sound = [noise_dict[i]*100 for i in l ]
            g_heuristic_1 = [get_distance(dict_of_nodes[pointer].on_road_coordinates,dict_of_nodes[i].on_road_coordinates) for i in l]
            g_heuristic = [i + cost for i in g_heuristic_1]
            g_heuristic_final = list(map(lambda x,y : x + y , g_heuristic,g_heuristic_sound))

            h_heuristic = [get_distance(dict_of_nodes[i].on_road_coordinates,dict_of_nodes[destination].on_road_coordinates) for i in l]
            f_heuristic = list(map(lambda x,y : x + y , g_heuristic_final,h_heuristic))
            f_list = list(zip(l,f_heuristic))
            sorted_list = sorted(f_list,key = lambda x: x[1])
            next_pointer = sorted_list[0][0]
            k = l.index(next_pointer)
            if dict_of_nodes[pointer].neighbours[next_pointer]!= 0 :
                set_of_coordinates.extend(dict_of_nodes[pointer].neighbours[next_pointer])

            elif dict_of_nodes[next_pointer].neighbours[pointer]!= 0 :
                r = dict_of_nodes[next_pointer].neighbours[pointer]
                r.reverse()
                set_of_coordinates.extend(r)
                
            else:
                set_of_coordinates.append(dict_of_nodes[pointer].on_road_coordinates)
                set_of_coordinates.append(dict_of_nodes[next_pointer].on_road_coordinates)

            pointer = next_pointer

            cost += g_heuristic_1[k]


   
        if set_of_coordinates[-1] != dict_of_nodes[destination].on_road_coordinates:
            set_of_coordinates.append(dict_of_nodes[destination].on_road_coordinates)

        # m = folium.Map(location= set_of_coordinates[0], zoom_start=13)

        #                                                # draw the route on the map
        # folium.PolyLine(set_of_coordinates, color='red', weight=2, opacity=0.7).add_to(m)

        return(cost,set_of_coordinates)
    
    if genetic == True:
        start,destination = (points[0],points[1])
        set_of_coordinates = []
        pointer = start
        cost = 0
        
        route_count = 0 
        
        global iteration_count
        global population
        g_3 = False
        iteration_count = 0
        population = [] 
        
        
        while len(population) <= 7 and iteration_count <= 2000:
            
            route = []
            route.append(start)
            pointer = start
            while route[-1]!= destination or len(route) <= 10:
                iteration_count +=1
                k = list(dict_of_nodes[pointer].neighbours.keys())
                k_2 = list(filter(lambda x: x not in route,k))
                if k_2 == []:
                    g_3 = True
                    break

                k_3 = random.choice(k_2)

                
                route.append(k_3)
                pointer = k_3
                

            if route in population:

                continue

            elif route[-1] == destination:
                population.append(route)
                route_count+=1
            
            else:

                continue

# Selection on the basis of fitness function
        global population_2
        global child_array
        global counter
        
        counter = 0
        population_2 = sorted(population,key = array_cost,reverse = True)
        if len(population_2) > 8 :
            
            
            population_2 = population_2[:8]
        
        
        child_array = population_2
        while len(child_array) > 1:

            parents_array = [(child_array[i],child_array[i+1]) for i in range(len(child_array)-1)]

            child_array = [reproduce(x[0],x[1]) for x in parents_array]

            child_array_2 = sorted(child_array,key = array_cost,reverse = True)

            child_array = child_array_2
            
            counter+=1


        return child_array[0]
            
#             
```
